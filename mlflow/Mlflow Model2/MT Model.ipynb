```python
import mlflow
mlflow.set_tracking_uri("http://localhost:5000")
```

### Model Tracking

```python

from transformers import pipeline

pipe = pipeline("text2text-generation", model="Messam174/mbart-large-50-many-to-many-mmt")


with mlflow.start_run():
    mlflow.transformers.log_model(
        transformers_model=pipe,
        artifact_path="model",
    )
```

### Model tracking with changing Hyperparameter

```python
# Import necessary libraries
from transformers import pipeline
import mlflow


pipe = pipeline("text2text-generation", model="Messam174/mbart-large-50-many-to-many-mmt")

# Define the hyperparameters
hyperparameters = {
    "learning_rate":2e-05,
    "train_batch_size": 4,
    "eval_batch_size": 4,
    "seed": 42,
    "gradient_accumulation_steps": 2,
    "total_train_batch_size": 8,
    "optimizer": "Adam with betas=(0.9, 0.999) and epsilon=1e-08",
    "lr_scheduler_type": "linear",
    "num_epochs": 10
}

# Start an MLflow run
with mlflow.start_run():
    # Log hyperparameters to MLflow
    mlflow.log_params(hyperparameters)

    # Log the Wav2Vec2 pipeline model with MLflow
    mlflow.transformers.log_model(
        transformers_model=pipe,
        artifact_path="model",
        registered_model_name="text2text-generation-model"  
    )

print("Model and hyperparameters successfully tracked.")
```
